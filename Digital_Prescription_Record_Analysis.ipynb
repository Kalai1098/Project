```python
import mysql.connector

def get_connection():

    return mysql.connector.connect(
        host="localhost",
        user="root",
        password="1098",
        database="digital_prescription_db"
    )
```

```python
def execute_query(query):

    conn = get_connection()

    cursor = conn.cursor()

    cursor.execute(query)

    result = cursor.fetchall()

    conn.close()

    return result
```

```python
def most_prescribed_medicines():

    query = """
    SELECT medicine_name,
           COUNT(*) AS total
    FROM prescriptions_details
    GROUP BY medicine_name
    ORDER BY total DESC
    LIMIT 10;
    """

    results = execute_query(query)

    print("\nTOP 10 PRESCRIBED MEDICINES\n")

    for medicine, total in results:
        print(f"{medicine:<25} {total}")
```

```python
def doctor_prescription_patterns():

    query = """
    SELECT doctor_name,
           COUNT(*) AS total
    FROM prescriptions_details
    GROUP BY doctor_name
    ORDER BY total DESC;
    """

    results = execute_query(query)

    print("\nDOCTOR PRESCRIPTION PATTERNS\n")

    for doctor, total in results:
        print(f"{doctor:<25} {total}")
```

```python
def specialty_trends():

    query = """
    SELECT doctor_specialty,
           COUNT(*) AS total
    FROM prescriptions_details
    GROUP BY doctor_specialty
    ORDER BY total DESC;
    """

    results = execute_query(query)

    print("\nSPECIALTY-WISE PRESCRIPTIONS\n")

    for specialty, total in results:
        print(f"{specialty:<20} {total}")
```

```python
def disease_medicine_mapping():

    query = """
    SELECT disease_name,
           medicine_name,
           COUNT(*) AS total
    FROM prescriptions_details
    GROUP BY disease_name,
             medicine_name
    ORDER BY total DESC
    LIMIT 20;
    """

    results = execute_query(query)

    print("\nDISEASE TO MEDICINE MAPPING\n")

    for disease, medicine, total in results:

        print(
            f"{disease:<20}"
            f"{medicine:<20}"
            f"{total}"
        )
```

```python
def prescription_duration():

    query = """
    SELECT medicine_name,
           ROUND(AVG(duration_days),2)
    FROM prescriptions_details
    GROUP BY medicine_name
    ORDER BY AVG(duration_days) DESC;
    """

    results = execute_query(query)

    print("\nAVERAGE PRESCRIPTION DURATION\n")

    for medicine, duration in results:

        print(
            f"{medicine:<25}"
            f"{duration} Days"
        )
```

```python
def prescription_trend():

    query = """
    SELECT MONTH(diagnosis_date),
           COUNT(*)
    FROM prescriptions_details
    GROUP BY MONTH(diagnosis_date)
    ORDER BY MONTH(diagnosis_date);
    """

    results = execute_query(query)

    print("\nMONTHLY PRESCRIPTION TREND\n")

    for month, total in results:

        print(
            f"Month {month}: {total}"
        )
```

```python
def repeat_prescriptions():

    query = """
    SELECT patient_id,
           COUNT(*) AS total
    FROM prescriptions_details
    GROUP BY patient_id
    HAVING COUNT(*) > 1
    ORDER BY total DESC
    LIMIT 20;
    """

    results = execute_query(query)

    print("\nREPEAT PRESCRIPTIONS\n")

    for patient, total in results:

        print(
            f"{patient:<15}"
            f"{total}"
        )
```

```python
def healthcare_cli():

    while True:

        print("\n")
        print("="*50)
        print("DIGITAL PRESCRIPTION RECORD ANALYSIS")
        print("="*50)

        print("1. Most Prescribed Medicines")
        print("2. Doctor-Wise Prescription Patterns")
        print("3. Specialty-Wise Prescription Trends")
        print("4. Disease-to-Medicine Mapping")
        print("5. Prescription Duration Analysis")
        print("6. Prescription Trend Over Time")
        print("7. Patient Retention & Repeat Prescriptions")
        print("8. Exit")

        choice = input("\nEnter Choice: ")

        if choice == "1":
            most_prescribed_medicines()

        elif choice == "2":
            doctor_prescription_patterns()

        elif choice == "3":
            specialty_trends()

        elif choice == "4":
            disease_medicine_mapping()

        elif choice == "5":
            prescription_duration()

        elif choice == "6":
            prescription_trend()

        elif choice == "7":
            repeat_prescriptions()

        elif choice == "8":
            print("\nThank You")
            break

        else:
            print("\nInvalid Choice")
```
